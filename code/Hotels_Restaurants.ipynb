```python
# Dependencies
import requests
import json
import pandas as pd

# Google developer API key
from config import gkey
```

```python
#path to csv file
file = "research_data/city_list.csv"

#read data into pandas
citydf = pd.read_csv(file)
citydf.head()
```

```python
# geocoordinates
# have to go through documentation to understand all the parameters
# target type is what is your type of search, restaurants, movie theaters, etc
target_coordinates = for lat in citydf:    
target_radius = 40000
target_type = "restaurant"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
# pass base url and then pass all parameters in the argument params
response = requests.get(base_url, params=params)
```
