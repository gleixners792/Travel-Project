```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://en.wikipedia.org/wiki/Cleveland'
browser.visit(url)
time.sleep(5)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
cities_file = "cities.csv"
cities_df = pd.read_csv(cities_file)
cities_df.head()
```

```python
cities_df.info()
```

```python
jpeg_imgs = []

for city, row in cities_df.iterrows():

    url = 'https://en.wikipedia.org/wiki/'+ row['Wiki_String']
    
    #print(url)
    browser.visit(url)
    time.sleep(3)

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    images = soup.find_all('img', {'src':re.compile('.jpg')})
    jpeg_img = images[0]['src']
    jpeg_imgs.append(jpeg_img)
    print(jpeg_img)

print(" ")
print("Done with the scraping")
print(" ")
jpeg_imgs
```

```python
cities_df['img_url'] = jpeg_imgs
cities_df.info()
```

```python
output_data_file = "cities_with_url.csv"
cities_df.to_csv(output_data_file, index= False)
```
